# Binning


Binning is a technique used for Data Smoothing. real world data stores noise along with the actual data. the noisy data is meaningless and also reduces the performance of the model, when not removed.

The data is first sorted and then the distribution is divided into the defined number of bins.

## Types of Binning

1. Equal Distance(Width) Binning 

```
The simplest binning appraoch is to divide the data into bins with equal number of elements.

This technique is best suited for Normalized/ Standardized data.

```



2. Equal Frequency (depth) Binning
```
In this technique, the number of elements in each bin are equal (approximately).

```

### Smoothing by Means


*   **Smoothing by Means** - Each value in the bin is replaced the mean of the bin.


````python
# Bin mean
for i in range (0,150,5):
    k=int(i/5)
    mean=(b[i] + b[i+1] + b[i+2] + b[i+3] + b[i+4])/5
    for j in range(5):
        bin1[k,j]=mean
print("Bin Mean: \n",bin1)
````

### Smoothing by Boundary values


*   **Smoothing by Boundary** - In this binning approach, the minimum and maximum value of the bin is identified, and then the values of the elements in the bin are replaced by the closest boundary value.

````python
# Bin boundaries
for i in range (0,150,5):
    k=int(i/5)
    for j in range (5):
        if (b[i+j]-b[i]) < (b[i+4]-b[i+j]):
            bin2[k,j]=b[i]
        else:
            bin2[k,j]=b[i+4]       
print("Bin Boundaries: \n",bin2)
````

### Smoothing by Median

*   **Smoothing by Median** - Each value in the bin is replaced the median of the bin.

````python
# Bin median
for i in range (0,150,5):
    k=int(i/5)
    for j in range (5):
        bin3[k,j]=b[i+2]
print("Bin Median: \n",bin3)
````

# The use of Binning

The most common instances of binning technique is used in the backend when  creating a Histogram. The histogram below of customer sales data, shows how a continuous set of sales numbers can be divided into discrete bins (for example: $60,000 - $70,000) and then used to group and count account instances.

````python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')

raw_df = pd.read_excel('https://github.com/chris1610/pbpython/blob/master/data/2018_Sales_Total_v2.xlsx?raw=true')
raw_df.head()
````

````python
df = raw_df.groupby(['account number', 'name'])['ext price'].sum().reset_index()
````

````python
df.head()
````

````python
df['ext price'].plot(kind='hist')
plt.show()
````

In the histogram above there are 8 bins in the graph.
But there are scenarios where one wants to define the number of bins as per your needs. Pandas cut() and qcut() functions are used for the same.


## qcut()

image.png

Pandas qcut() function is the Percentile based Discretization Function.
qcut tries to divide the data into Equal sized bins.
The function defines the bins using percentiles based on the distribution of the data, not the actual numeric edges of the bins.

The simplest use of qcut is to define the number of quantiles and let pandas figure out how to divide up the data

````python
pd.qcut(df['ext price'], q=4)
````

The categories at the bottom of the output above shows how to divide up the data into 4 bins. 

**Bin1** = (55733.049000000006, 89137.708],

**Bin 2** = (89137.708, 100271.535], 

**BIn 3** = (100271.535, 110132.552],

**BIn 4** = (110132.552, 184793.7]

````python
# adding a new column with the bin value
df['qcut_4'] = pd.qcut(df['ext price'], q=4, precision=2)   # precision for number of decimals.
df.head()
````

````python
# let us add another coulumn with bin=10
df['qcut_10'] = pd.qcut(df['ext price'], q=10, precision =2)
df.head()
````

````python
# let us check the frequency in each bin for quantiles
df['qcut_4'].value_counts()
````

````python
# let us check the frequency in each bin for deciles
df['qcut_10'].value_counts()
````

As you can clearly see that pandas divided the complete dataframe into rqual number of bins. As python does the maths behind the scenes for us, it is a bit confusing to explain this to someone.

When you see the range of each bin, you will find a variation in range for each individual bin. 

Try to find out the same. Simply subtract the lower limit of the bin from the upper limit. 





**Let us now label the bins for easy identification.**

````python
bin_labels_5 = ['Bronze', 'Silver', 'Gold', 'Platinum', 'Diamond']
df['qcut_title'] = pd.qcut(df['ext price'],
                              q=[0, .2, .4, .6, .8, 1],
                              labels=bin_labels_5)
df.head()
````
